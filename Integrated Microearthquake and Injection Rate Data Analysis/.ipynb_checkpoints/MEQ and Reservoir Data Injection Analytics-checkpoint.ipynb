# Integrated MEQ and Injection Rate Data Analysis

### By Geophysics Team PT Supreme Energy

Notebook ini berisikan analisis data terintegrasi microearthquake (MEQ) dengan laju injeksi. Analisis data terintegrasi ini diperlukan untuk menemukan pola dan wawasan dari fitur-fitur yang tersedia baik dari katalog MEQ dan data injeksi fluida untuk kebutuhan monitoring produksi geothermal. Injeksi fluida pada bawah permukaan mampu menstimulasi terjadinya perekahan batuan, sehingga menghasilkan seismisitas terinduksi. Hal tersebut akan mengakibatkan terjadinya perambatan gelombang gempa mikro yang akan direkam oleh seismometer. 

Analisis data menggunakan pendekatan data science dilakukan pada data reservoir sumur-sumur injeksi yakni B1, C1, D1, D2, E1, dan E2 dari Januari 2021 hingga Mei 2023 digabungkan dengan data katalog hipocenter dari December 2020 hingga Oktober 2023.

### Data Understanding

- `Time_Record_ML_Injection_Well`: tanggal kegiatan injeksi dan diperolehnya kejadian gempa mikro
- `Well Status`: Status well pada tanggal tersebut, baik itu sedang melakukan injeksi (injection) atau sedang dihentikan (shut-in)
- `Injection Rate`: laju injeksi fluida dalam unit (l/s)
- `Total_MEQ_Event`: jumlah kejadian gempa mikro pada hari tertentu

```python
data.describe(include=['O'])
```

* Sumur injeksi ML D1 dan ML E2 cenderung tidak aktif sebab didominasi unique value 'shut-in'.
* Seharusnya unique value disetiap kolom well status hanya 2, yakni 'injection' dan 'shut-in' akan tetapi dibeberapa fitur terlihat lebih dari dua unique value, hal ini perlu kita cek karena bisa jadi terdapat missing value atau terdapat unique value yang merepresentasikan hal yang sama, namun bisa jadi karena adanya ketidak-konsistenan penulisan sehingga memiliki lebih dari dua unique value.

```python
categoric_data = data.select_dtypes(include="object")

column = categoric_data.columns
unique = [categoric_data[i].unique() for i in column]
n_unique = [len(categoric_data[i].unique()) for i in column]

pd.DataFrame({
    'column': column,
    'n_unique': n_unique,
    'unique': unique
})
```

* Dari informasi table unique value diatas terlihat terjadinya ketidak-konsistenan dalam penulisan, sehingga akan diperbaiki untuk kedepannya hanya ada dua unique value yakni 'Injecting' dan 'Shut-in'.

```python
data["Well Status ML B1"].replace(to_replace=["Shut-In", "shut-in"], value="Shut-in", regex=True, inplace=True)
data["Well Status ML C1"].replace(to_replace=["Shut-In", "shut-in"], value="Shut-in", regex=True, inplace=True)
data["Well Status ML D1"].replace(to_replace=["Shut-In", "shut-in"], value="Shut-in", regex=True, inplace=True)
data["Well Status ML D2"].replace(to_replace=["Shut-In", "shut-in"], value="Shut-in", regex=True, inplace=True)
data["Well Status ML E2"].replace(to_replace=["Shut-In", "shut-in"], value="Shut-in", regex=True, inplace=True)
```

```python
categoric_data = data.select_dtypes(include="object")

column = categoric_data.columns
unique = [categoric_data[i].unique() for i in column]
n_unique = [len(categoric_data[i].unique()) for i in column]

pd.DataFrame({
    'column': column,
    'n_unique': n_unique,
    'unique': unique
})
```

Setelah ketidak-konsistenan penulisan pada dataset diperbaiki, sekarang akan dimulai pengisian missing value.

```python
# List nama kolom well status
well_status_columns = [
    "Well Status ML B1",
    "Well Status ML C1",
    "Well Status ML D1",
    "Well Status ML D2",
    "Well Status ML E1",
    "Well Status ML E2"
]

# List nama kolom injection rate
injection_rate_columns = [
    "Injection Rate ML B1 (l/s)",
    "Injection Rate ML C1 (l/s)",
    "Injection Rate ML D1 (l/s)",
    "Injection Rate ML D2 (l/s)",
    "Injection Rate ML E1 (l/s)",
    "Injection Rate ML E2 (l/s)"
]

# Loop melalui setiap kolom well status dan injection rate
for status_col, rate_col in zip(well_status_columns, injection_rate_columns):
    # Filter dataframe berdasarkan kolom well status
    data_injecting = data[data[status_col] == "Injecting"]
    data_shutin = data[data[status_col] == "Shut-in"]

    # Isi missing value pada kolom injection rate dengan 0
    data_shutin[rate_col] = 0

    # Gabungkan dataframe yang sudah difilter dengan dataframe asli
    data = data_injecting.append(data_shutin)
```

```python
data
```

```python
data = data.sort_index()
data
```

```python
missingno.matrix(data)
```

```python
for feature in injection_rate_columns:
    dp_statistics, df_pvalue = normaltest(data[feature].dropna())
    
    if df_pvalue > 0.05:
        print(f"{feature}: P-value: {df_pvalue}. We assume a normal distribution.")
    else:
        print(f"{feature}: P-value: {df_pvalue}. We do not assume a normal distribution.")
```

* Semua fitur numerikal laju injeksi memiliki distribusi nilai yang tidak normal, sehingga pengisian missing value akan menggunakan median pada data dari fitur tersebut.

```python
# Select float columns
float_columns = data.select_dtypes(include=[float]).columns

# Simple imputation using median for all float features
simple_impute = SimpleImputer(strategy='median')
data[float_columns] = simple_impute.fit_transform(data[float_columns])
```

```python
missingno.matrix(data)
```

* Penanganan missing value selesai, dan fitur yang tidak diperlukan di-drop, kemudian dataset siap untuk masuk ke proses analisis data terintegrasi.

```python
data.describe(include=['O'])
```

* Sumur ML-D1 dan ML-E2 merupakan sumur yang paling tidak aktif sebab didominasi oleh kondisi 'shut-in'
* Adapun plot laju injeksi dan total kejadian MEQ hanya akan menggunakan data dari fitur-fitur numerikal, sehingga fitur-fitur kategorikal akan di drop.

```python
# Drop fitur yang tidak diinginkan
data_unwanted_columns = [
    'Well Status ML B1',
    'Well Status ML C1',
    'Well Status ML D1',
    'Well Status ML D2',
    'Well Status ML E1',
    'Well Status ML E2'
]

data = data.drop(columns=data_unwanted_columns)
```

# Induced Seismicity and Reservoir Data Scraping

```python
df2 = pd.read_excel('Total_MEQ_Events.xlsx') 
df2
```

```python
df2['cum_sum_meq_event'] = df2['total_meq_events'].cumsum()
df2
```

```python
result_merge = pd.merge(data, df2, on='date')
result_merge
```

```python
result_merge.describe()
```

* Laju injeksi pada sumur C cenderung kecil (maksimum 9.3 l/s) sehingga sulit untuk menstimulasi terjadinya seismisitas terinduksi, maka dari itu kecil kemungkinan terjadinya perekahan akibat laju injeksi dari sumur C.
* Sumur ML C1 memiliki rata-rata laju injeksi paling rendah dibandingkan sumur injeksi lainnya.
* Sumur ML D2 sebagai sumur dengan rata-rata laju injeksi tertinggi, adapun dari Januari 2021 sampai Mei 2023 laju injeksi tertinggi ada pada sumur ML D2 dengan 250 (l/s)

```python
# Create a figure and axis for the first plot
fig, ax1 = plt.subplots(figsize=(20, 9))

# Plot the first set of data (Injection Rate) on the first y-axis (ax1)
ax1.plot(result_merge['date'], result_merge['Injection Rate ML B1 (l/s)'], color='green', label='Injection Rate ML B1')
ax1.plot(result_merge['date'], result_merge['Injection Rate ML C1 (l/s)'], color='orange', label='Injection Rate ML C1')
ax1.plot(result_merge['date'], result_merge['Injection Rate ML D1 (l/s)'], color='yellow', label='Injection Rate ML D1')
ax1.plot(result_merge['date'], result_merge['Injection Rate ML D2 (l/s)'], color='blue', label='Injection Rate ML D2')
ax1.plot(result_merge['date'], result_merge['Injection Rate ML E1 (l/s)'], color='pink',label='Injection Rate ML E1')
ax1.plot(result_merge['date'], result_merge['Injection Rate ML E2 (l/s)'], color='grey',label='Injection Rate ML E2')

# Set the label and color for the first y-axis
ax1.set_xlabel('Date')
ax1.set_ylabel('Injection Rate (l/s)', color='black')

# Create a second y-axis for the second plot
ax2 = ax1.twinx()

# Plot the second set of data (Total MEQ Event) on the second y-axis (ax2)
ax2.plot(result_merge['date'], result_merge['total_meq_events'], linewidth=0.75, marker='o', linestyle='--', color='black', label='Total MEQ Event')

# Set the label and color for the second y-axis
ax2.set_ylabel('Total MEQ Event')

# Combine legends from both y-axes
lines, labels = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax1.legend(lines + lines2, labels + labels2, loc='upper left')

# Set the title and adjust layout
#plt.title('Injection Rate vs. Total MEQ Event Over Time')
plt.xticks(rotation=45)
plt.tight_layout()

# Show the plot
plt.show()
```

* Plot diatas menunjukan total gempa mikro dan laju injeksi sumur ML B1, D1, D2, E1, dan E2 dari Januari 2021 sampai Mei 2023, terdapat beberapa insight yakni:
    1. Terjadi peningkatan total gempa mikro pada kuartal awal 2023 secara signifikan, meskipun laju injeksi pada sumur ML-B1, ML-D2, dan ML-E1 memiliki tren yang relatif konstan dari sebelum kejadian kenaikan total gempa mikro tersebut. 
    2. Tidak terjadi peningkatan total gempa mikro ketika sumur ML D2, E1, dan E2 shut-in diwaktu menjelang masuk bulan November 2021

```python
# Ekstrak informasi tahun
result_merge['year'] = result_merge['date'].dt.year
result_merge['month'] = result_merge['date'].dt.month
result_merge['day'] = result_merge['date'].dt.day
```

```python
result_merge.columns
```

```python
result_merge['Cum Sum Injection Rate ML B1 (l/s)'] = result_merge['Injection Rate ML B1 (l/s)'].cumsum()
result_merge['Cum Sum Injection Rate ML C1 (l/s)'] = result_merge['Injection Rate ML C1 (l/s)'].cumsum()
result_merge['Cum Sum Injection Rate ML D1 (l/s)'] = result_merge['Injection Rate ML D1 (l/s)'].cumsum()
result_merge['Cum Sum Injection Rate ML D2 (l/s)'] = result_merge['Injection Rate ML D2 (l/s)'].cumsum()
result_merge['Cum Sum Injection Rate ML E1 (l/s)'] = result_merge['Injection Rate ML E1 (l/s)'].cumsum()
result_merge['Cum Sum Injection Rate ML E2 (l/s)'] = result_merge['Injection Rate ML E2 (l/s)'].cumsum()
```

```python
# Create a figure and axis for the first plot
fig, ax1 = plt.subplots(figsize=(20, 9))

# Plot the first set of data (Injection Rate) on the first y-axis (ax1)
ax1.plot(result_merge['date'], result_merge['Cum Sum Injection Rate ML B1 (l/s)'], color='green', label='Cum. Sum Injection Rate ML B1')
ax1.plot(result_merge['date'], result_merge['Cum Sum Injection Rate ML C1 (l/s)'], color='orange', label='Cum. Sum Injection Rate ML C1')
ax1.plot(result_merge['date'], result_merge['Cum Sum Injection Rate ML D1 (l/s)'], color='yellow', label='Cum. Sum Injection Rate ML D1')
ax1.plot(result_merge['date'], result_merge['Cum Sum Injection Rate ML D2 (l/s)'], color='blue', label='Cum. Sum Injection Rate ML D2')
ax1.plot(result_merge['date'], result_merge['Cum Sum Injection Rate ML E1 (l/s)'], color='pink',label='Cum. Sum Injection Rate ML E1')
ax1.plot(result_merge['date'], result_merge['Cum Sum Injection Rate ML E2 (l/s)'], color='grey',label='Cum. Sum Injection Rate ML E2')

# Set the label and color for the first y-axis
ax1.set_xlabel('Date')
ax1.set_ylabel('Cumulative Sum Injection Rate (l/s)', color='black')

# Create a second y-axis for the second plot
ax2 = ax1.twinx()

# Plot the second set of data (Total MEQ Event) on the second y-axis (ax2)
ax2.plot(result_merge['date'], result_merge['cum_sum_meq_event'], linewidth=0.75, marker='o', linestyle='--', color='black', label='Cum. Sum MEQ Event')

# Set the label and color for the second y-axis
ax2.set_ylabel('Cum. Sum MEQ Event')

# Combine legends from both y-axes
lines, labels = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax1.legend(lines + lines2, labels + labels2, loc='upper left')

# Set the title and adjust layout
#plt.title('Injection Rate vs. Total MEQ Event Over Time')
plt.xticks(rotation=45)
plt.tight_layout()

# Show the plot
plt.show()
```

* Hanya terjadi peningkatan jumlah kumulatif gempa mikro secara signifikan pada awal kuartal pertama tahun 2023, meskipun semua sumur injeksi secara gradual meningkatkan laju injeksi dari waktu kewaktu
* Sumur ML-D2 merupakan sumur dengan jumlah kumulatif laju injeksi tertinggi dibandingkan sumur lainnya dari Januari 2021 sampai Mei 2023, disusul oleh sumur ML-E1. 
* Sumur ML-C1 merupakan sumur dengan jumlah kumulatif laju injeksi paling rendah dibandingkan sumur lainnya dari Januari 2021 sampai Mei 2023.
